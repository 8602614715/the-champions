```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('tempecture')
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('tempecture')
plt.title('Weather')
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label ='Min')
plt.plot(days,avg_t,label ='avg')
plt.legend()
```

```python
import numpy as np
```

```python
company = ['GOOGLE','FACEBOOK','AMAZON','INSTA']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
ypos = np.arange(len(company))
ypos
plt.bar(company,revenue)
```

```python
plt.xticks(ypos,company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('Data of companies')
plt.bar(company,revenue)
```

```python
plt.xticks(ypos,company)

plt.ylabel('Revenue')
plt.title('Data of companies')
plt.bar(ypos,revenue,label="Revenue")
plt.bar(ypos,profit,label="Profit")
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)

plt.ylabel('Revenue')
plt.title('Data of companies')
plt.bar(xpos-0.2,revenue,label="Revenue")
plt.bar(xpos+0.2,profit,label="Profit")
plt.legend()
```

```python
# control width of chart use width fxn
plt.xticks(ypos,company)

plt.ylabel('Revenue')
plt.title('Data of companies')
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
plt.yticks(ypos,company)
plt.title('Data of companies')
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = [80,100,125,150],rwidth=0.8)
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67,98,120,133,150,84,69,89,79,120,112,100]
plt.hist([blood_sugar_men,blood_sugar_women],bins = [80,100,125,150],rwidth= 0.95,color = ['green','red'],label = ['men','women'])
plt.legend()
```
